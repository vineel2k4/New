```python
import random

# Function to initialize the deck
def initialize_deck():
    colors = ["Red", "Yellow", "Green", "Blue"]
    values = [str(i) for i in range(10)] * 2
    special_cards = ["Skip", "Reverse", "Draw Two"] * 2
    wild_cards = ["Wild", "Wild Draw Four"] * 4
    deck = [(color, value) for color in colors for value in values]
    deck += [(color, special_card) for color in colors for special_card in special_cards]
    deck += [(wild_card, "Wild") for wild_card in wild_cards]
    return deck

# Function to shuffle the deck
def shuffle_deck(deck):
    random.shuffle(deck)
    return deck

# Function to deal cards to players
def deal_cards(deck, num_players):
    hands = [[] for _ in range(num_players)]
    for _ in range(7):
        for hand in hands:
            card = deck.pop(0)
            hand.append(card)
    return hands, deck

# Function to check if a card can be played
def can_play(card, current_card):
    return card[0] == current_card[0] or card[1] == current_card[1]

# Function to simulate Uno game
def uno_game(num_players):
    deck = initialize_deck()
    deck = shuffle_deck(deck)
    hands, deck = deal_cards(deck, num_players)
    current_card = deck.pop(0)
    player_turn = 0
    direction = 1
    game_over = False

    while not game_over:
        print("\nCurrent card:", current_card)
        print("Player", player_turn + 1, "'s turn")
        print("Your hand:", hands[player_turn])

        valid_card = False
        for card in hands[player_turn]:
            if can_play(card, current_card):
                valid_card = True
                print("Play", card)
                chosen_card = hands[player_turn].pop(hands[player_turn].index(card))
                current_card = chosen_card
                if not hands[player_turn]:
                    game_over = True
                break

        if not valid_card:
            print("No valid card. Drawing a card.")
            hands[player_turn].append(deck.pop(0))

        if not game_over:
            player_turn = (player_turn + direction) % num_players

    print("Player", player_turn + 1, "wins!")

# Start the game with 4 players
uno_game(4)
```
